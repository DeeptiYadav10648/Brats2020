```python
from src.datasets.brats_dataset import create_file_list
from src.transforms.preprocessing import (
    train_transform,
    val_transform
)
from torch_geometric.loader import DataLoader
```

```python
from src.config import DATA_DIR

all_files = create_file_list(DATA_DIR)

print(len(all_files))
```

```python
from sklearn.model_selection import train_test_split

train_files, val_files = train_test_split(
    all_files,
    test_size=0.2,
    random_state=42
)

print(len(train_files))
print(len(val_files))
```

```python
from monai.data import Dataset

from src.transforms.preprocessing import (
    train_transform,
    val_transform
)

train_dataset = Dataset(
    data=train_files,
    transform=train_transform
)

val_dataset = Dataset(
    data=val_files,
    transform=val_transform
)
```

```python
from src.graphs.create_graph_dataset import create_graph_dataset

print("Creating training graphs...")
train_graphs = create_graph_dataset(train_dataset)

print("Creating validation graphs...")
val_graphs = create_graph_dataset(val_dataset)
```

```python
from torch_geometric.loader import DataLoader

train_loader = DataLoader(
    train_graphs,
    batch_size=1,
    shuffle=True
)

val_loader = DataLoader(
    val_graphs,
    batch_size=1,
    shuffle=False
)
```

```python
from train_isgnn import train_isgnn

model = train_isgnn(
    train_loader,
    val_loader,
    epochs=300,
    lr=1e-3,
    hidden=128,
    patience=40
)
```
